```python
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>I Love You - Explosion</title>
  <style>
    body {
      display: flex;
      justify-content: center;
      align-items: center;
      height: 100vh;
      background-color: #f8f8f8;
      margin: 0;
      font-family: 'Arial', sans-serif;
      overflow: hidden;
    }

    .heart {
      position: relative;
      width: 200px;
      height: 200px;
      background-color: red;
      border-radius: 50% 50% 0 0;
      transform: rotate(45deg);
      cursor: pointer;
      animation: pulse 1s infinite alternate;
    }

    .heart:before,
    .heart:after {
      content: '';
      position: absolute;
      width: 200px;
      height: 200px;
      border-radius: 50%;
      background-color: red;
    }

    .heart:before {
      left: 0;
      top: -100px;
    }

    .heart:after {
      left: 100px;
      top: 0;
    }

    .message {
      position: absolute;
      top: 50%;
      left: 50%;
      transform: translate(-50%, -50%);
      font-size: 24px;
      color: white;
      font-weight: bold;
      text-shadow: 2px 2px 10px rgba(0, 0, 0, 0.6);
      pointer-events: none;
    }

    @keyframes pulse {
      0% {
        transform: scale(1);
      }
      100% {
        transform: scale(1.1);
      }
    }

    .explode {
      position: absolute;
      top: 50%;
      left: 50%;
      transform: translate(-50%, -50%);
      font-size: 16px;
      color: red;
      font-weight: bold;
      opacity: 1;
      animation: explodeAnimation 2s forwards;
    }

    @keyframes explodeAnimation {
      0% {
        transform: translate(-50%, -50%) scale(1);
        opacity: 1;
      }
      100% {
        transform: translate(-50%, -200%) scale(3);
        opacity: 0;
      }
    }
  </style>
</head>
<body>

  <div class="heart" onclick="explodeHeart(event)">
    <div class="message">I LOVE YOU</div>
  </div>

  <script>
    function explodeHeart(event) {
      // Get the heart position
      const heart = event.target;

      // Generate a "I LOVE YOU" explosion
      for (let i = 0; i < 30; i++) {
        const explodeText = document.createElement('div');
        explodeText.classList.add('explode');
        explodeText.textContent = "I LOVE YOU";

        // Randomly position and animate each explosion piece
        const angle = Math.random() * Math.PI * 2;
        const distance = Math.random() * 300;
        const offsetX = Math.cos(angle) * distance;
        const offsetY = Math.sin(angle) * distance;
        const duration = Math.random() * 1.5 + 1; // Random duration for a more natural effect

        explodeText.style.animationDuration = `${duration}s`;

        // Append to the body
        document.body.appendChild(explodeText);

        // Set random positions for the explosion
        explodeText.style.left = `${event.clientX}px`;
        explodeText.style.top = `${event.clientY}px`;

        // Remove the explosion piece after animation ends
        setTimeout(() => {
          explodeText.remove();
        }, duration * 1000);
      }
    }
  </script>

</body>
</html>
```
